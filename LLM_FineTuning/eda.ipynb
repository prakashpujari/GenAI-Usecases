### Cell 19: Filter One Category
Input: DataFrame `df` and target label (`tech`).
Process: Filter rows where `category == target_label`.

```python
target_label = 'tech'
target_df = df[df['category']== target_label]
```

### Cell 20: Build Token Frequency
Input: Text values from `target_df['text']`.
Process: Tokenize, lowercase, remove stopwords/non-alphanumeric tokens, count with `Counter`.
Output: Printed token-frequency mapping.

```python
# Choose the label for which you want to create the word cloud
stop_words = set(stopwords.words('english'))
target_words = [word.lower() for text in target_df['text']
                for word in word_tokenize(text)
                if word.isalnum() and word.lower() not in stop_words]

target_word_counts = Counter(target_words)
print(target_word_counts)
```

### Cell 21: Generate Word Cloud Data
Input: Token frequency dictionary `target_word_counts`.
Process: Build `WordCloud` object from frequencies.
Output: No direct display; word cloud object is stored in `word_cloud`.

```python
word_cloud = WordCloud().generate_from_frequencies(target_word_counts)
```

### Cell 22: Display Word Cloud
Input: Word cloud object `word_cloud`.
Process: Render image with matplotlib and set title.
Output: Visual word cloud plot for the selected category.

```python
plt.imshow(word_cloud,interpolation='bilinear')
plt.axis("off")
plt.title(f"World cloud for category: {target_label}")
plt.show()
```

### Cell 24: Define Token Counting Functions
Input: Text string(s), DataFrame, model encoding name.
Process: Define reusable functions to count tokens per row and across a DataFrame.
Output: No direct display; functions are available for subsequent cells.

```python
def cal_num_tokens_from_row(string: str, encoding_name: str) -> int:
    encoding = tiktoken.encoding_for_model(encoding_name)
    token_count = len(encoding.encode(string))
    return token_count

def cal_num_tokens_from_df(df, encoding_name: str) -> int:
    total_tokens = 0
    for text in df['text']:
        total_tokens += cal_num_tokens_from_row (text, encoding_name)
    return total_tokens
```

### Cell 25: Count Tokens in Full Dataset
Input: Cleaned DataFrame `df` and encoding `gpt-3.5-turbo`.
Process: Aggregate token counts across all text rows.
Output: Printed total token count for full dataset.

```python
total_tokens = cal_num_tokens_from_df(df, "gpt-3.5-turbo")
print(f"total numbver of tokens in the dataframe: {total_tokens}")
```

### Cell 28: Build Text/Label Records
Input: DataFrame `df` columns `text` and `category`.
Process: Build list of dictionaries with normalized `text` and mapped `label`.

```python
rows = [{'text': row['text'].strip(),
                    'label': row['category'],
                    } for idx, row in df.iterrows()]
```

### Cell 29: Shuffle Records
Input: List `rows`.
Process: Set random seed and shuffle rows for reproducible split order.
Output: No direct display; `rows` order is randomized.

```python
random.seed(42)
random.shuffle(rows)
```

### Cell 30: Define Train/Test Split
Input: Shuffled `rows` and `num_test = 500`.
Process: Slice into `test` and `train` partitions.
Output: No direct display; dictionary `splits` is created.

```python
num_test = 500
splits = {'test':rows[0:num_test], 'train':rows[num_test:]}
```

### Cell 31: Write Split CSV Files
Input: `splits` dictionary and output folder `dataset`.
Process: Write `dataset/train.csv` and `dataset/test.csv` with headers.
Output: No direct display; two CSV files are generated.

```python
for split in ['train', 'test']:
    with open(f'dataset/{split}.csv', 'w', newline='') as fOut:
        writer = csv.DictWriter(fOut, fieldnames=['text', 'label'])
        writer.writeheader()
        for row in splits[split]:
            writer.writerow(row)
```

### Cell 34: Export Fine-Tuning JSONL
Input: `final_df` rows and system instruction string.
Process: Convert each row into OpenAI chat fine-tuning message format and write JSONL.
Output: No direct display; file `test.jsonl` is created.

```python
system = "You are a intelligent assistant designed to classify news articles into these categories: business, entertainment, politics, sport, tech"
with open("test.jsonl", "w") as outfile:
    for _, row in final_df.iterrows():
        openai_format = {
            "messages": [
                {"role": "system", "content": system},
                {"role": "user", "content": row["text"]},
                {"role": "assistant", "content": row["label"]}
            ]
        }
        json.dump(openai_format, outfile)
        outfile.write('\n')    
```
